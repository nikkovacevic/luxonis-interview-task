```python
import pandas as pd

df = pd.read_csv("appartments.csv", sep=",", index_col=0)
df.head()
```

```python
df.drop('web-scraper-start-url', axis=1, inplace=True)
df.drop('pagination', axis=1, inplace=True)
df.drop('appartment-link', axis=1, inplace=True)

df.head()
```

```python
df.isnull().sum()
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.drop(df.tail(205).index, inplace=True)
df.shape
```

```python
df.rename(columns={'web-scraper-order': 'id', 'appartment-link-href': 'link', 'image-src': 'image'}, inplace=True)
df.index.names = ['id']
df.head()
```

```python
df.to_csv('data.csv', header=False, index=False, encoding='utf-8-sig')
```
